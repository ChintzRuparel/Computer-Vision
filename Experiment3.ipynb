```python
# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import cv2
```

```python
# Reading the image
img= imread('puppy.jpeg')
plt.imshow(img)
img.shape
```

```python
#resizing the image to 128*64
img_r=resize(img,(128,64))
```

```python
#hog will return features and visualization of the image
feature,img_v=hog(img_r,orientations=9,pixels_per_cell=(4,4),cells_per_block=(2,2),visualize=True,multichannel=True)
feature2,img_v2=hog(img_r,orientations=9,pixels_per_cell=(32,32),cells_per_block=(2,2),visualize=True,multichannel=True)
feature3,img_v3=hog(img_r,orientations=9,pixels_per_cell=(8,8),cells_per_block=(8,8),visualize=True,multichannel=True)
```

```python
feature.shape
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(img_r,cmap='gray')
plt.title('Resized Image')
plt.subplot(1,3,3)
plt.imshow(img_v,cmap='gray')
plt.title('Visualization of the image')
```

```python
#plotting the images
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(img_v,cmap='gray')
plt.title('PPC=4*4 and CPB=2*2')
plt.subplot(1,3,2)
plt.imshow(img_v2,cmap='gray')
plt.title('PPC=32*32 and CPB=2*2')
plt.subplot(1,3,3)
plt.imshow(img_v3,cmap='gray')
plt.title('PPC=8*8 and CPB=8*8')
```

#Conclusion

Histogram of gradient for the given image is computed using the following parameters

1. Pixels per cell = 4X4, Number of cells per block = 2X2

2. Pixels per cell = 32X32, Number of cells per block = 2X2

3. Pixels per cell = 8X8, Number of cells per block = 8X8

For the least values of pixels per cell and number of cells per block, HOG of the image shows fine details of the edges

For the Size 32x32 the resized image shows 4 rows and 4 columns of cells. Doesnt identify image as cell sizes is larger then size of objects.

For the 3rd category, It gives you optimum number of combinations and shows and detects of the animal.

If resizing of image is done to include all rows and columns of the orignal image HOG detects all the edges of the given image.
