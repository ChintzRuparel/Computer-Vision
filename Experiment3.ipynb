```python
# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import cv2
```

```python
# Reading the image
img= imread('puppy.jpeg')
plt.imshow(img)
img.shape
```

```python
#resizing the image to 128*64
img_r=resize(img,(128,64))
```

```python
#
feature,img_v=hog(img_r,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2),visualize=True,multichannel=True)
```

```python
feature.shape
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(img_r,cmap='gray')
plt.title('Resized Image')
plt.subplot(1,3,3)
plt.imshow(img_v,cmap='gray')
plt.title('Visualization of the image')
```
