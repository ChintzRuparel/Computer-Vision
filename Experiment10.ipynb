```python
#Importing required libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading the required images
img_l= cv2.imread("left_image1.JPG")
img_r= cv2.imread("right_image1.JPG")
```

```python
#Converting Color image to gray
img_left = cv2.cvtColor(img_l, cv2.COLOR_RGB2GRAY)
img_right = cv2.cvtColor(img_r, cv2.COLOR_RGB2GRAY)
```

```python
#Shape of the images
sh1= img_left.shape
sh2= img_right.shape
```

```python
#Resizing right image
img_right_g= cv2.resize(img_right, (sh1[1], sh1[0]))
```

```python
img_right_g.shape
```

```python
sh1
```

```python
stereo1 = cv2.StereoBM_create(numDisparities=48, blockSize=31)   #numdisparitites sets the depth to be measured
depth1= stereo1.compute(img_left, img_right_g, cv2.CV_8UC1)  #Blocksize determines the size of the neighbourhood area
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 3, 1)
plt.imshow(img_l)
plt.title('Left Image')

plt.subplot(1, 3, 2)
plt.imshow(img_r)
plt.title('Right Image')

plt.subplot(1, 3, 3)
plt.imshow(depth1, cmap='gray')
plt.title('Depth Image with Disparity= 48 & Blocksize=31')
```

```python
#Comparison of disparities 16, 32 & 48
stereo2 = cv2.StereoBM_create(numDisparities=16, blockSize=31)
stereo3 = cv2.StereoBM_create(numDisparities=32, blockSize=31)
stereo4 = cv2.StereoBM_create(numDisparities=48, blockSize=31)
depth2= stereo2.compute(img_left, img_right_g, cv2.CV_8UC1)
depth3= stereo3.compute(img_left, img_right_g, cv2.CV_8UC1)
depth4= stereo4.compute(img_left, img_right_g, cv2.CV_8UC1)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 3, 1)
plt.imshow(depth2, cmap='gray')
plt.title('Disparity = 16')

plt.subplot(1, 3, 2)
plt.imshow(depth3, cmap='gray')
plt.title('Disparity = 32')

plt.subplot(1, 3, 3)
plt.imshow(depth4, cmap='gray')
plt.title('Disparity = 48')
```

```python
#Comparison of blocksizes 11, 21, 31 & 41
stereo5 = cv2.StereoBM_create(numDisparities=48, blockSize=11)
stereo6 = cv2.StereoBM_create(numDisparities=48, blockSize=21)
stereo7 = cv2.StereoBM_create(numDisparities=48, blockSize=31)
stereo8 = cv2.StereoBM_create(numDisparities=48, blockSize=41)
depth5= stereo5.compute(img_left, img_right_g, cv2.CV_8UC1)
depth6= stereo6.compute(img_left, img_right_g, cv2.CV_8UC1)
depth7= stereo7.compute(img_left, img_right_g, cv2.CV_8UC1)
depth8= stereo8.compute(img_left, img_right_g, cv2.CV_8UC1)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 4, 1)
plt.imshow(depth5, cmap='gray')
plt.title('Blocksize = 11')

plt.subplot(1, 4, 2)
plt.imshow(depth6, cmap='gray')
plt.title('Blocksize = 21')

plt.subplot(1, 4, 3)
plt.imshow(depth7, cmap='gray')
plt.title('Blocksize = 31')

plt.subplot(1, 4, 4)
plt.imshow(depth8, cmap='gray')
plt.title('Blocksize = 41')
```

#Conclusion
1. Depth Image is determined using Stereo images for disparity = 48 & blocksize =31.
2. Depth Map shows that objects which are near the camera have larger intensity than the objects that are far from the camera.
3. In the given image, person is nearest and therefore the pixels are white.
4. To get more details of depth of shadow of the object or small objects, the blocksize is reduced from 31 to 11.
5. To get broad details for the complete object like human from the image, the block size is increased 41.
6. To get depth of only objects that are near the camera, number of disparities is reduced from 48 to 16. Hence, the depth map shows fewer objects as disparity of 16 is not enough to capture all the objects in the image.
